# **Q2) List the five people with least posts and reactions combined. (least active people**

**Reading of the posts table for better understanding**

```python
posts_df
```

**Counting of the user id and the number of posts done by them in terms of user id**

```python
posts_user_count_df = posts_df[["User"]].copy()
posts_user_count_df = pd.DataFrame(posts_user_count_df.groupby(["User"]).size())
posts_user_count_df = posts_user_count_df.reset_index()
posts_user_count_df.columns = ["user_id", "posts_count"]
posts_user_count_df
```

**Reaction table calling here for better understanding of the data**

```python
reactions_df
```

**Reactions count for individual user id and displaying it in the coloumn of "reaction count"**

```python
reactions_user_count_df = reactions_df[["User"]].copy()
reactions_user_count_df = pd.DataFrame(reactions_user_count_df.groupby(["User"]).size())
reactions_user_count_df = reactions_user_count_df.reset_index()
reactions_user_count_df.columns = ["user_id", "reaction_count"]
reactions_user_count_df
```

**Now merging of the user dataframe and the reaction data frame to find individual person reaction count with the name**

```python
user_posts_reactions_df = user_df.merge(reactions_user_count_df,
                                        on = 'user_id',
                                        how="left")
user_posts_reactions_df
```

```python
user_df
```

**Merging of the Posts count reaction count and user data frame together** 

```python
user_posts_reactions_df = user_posts_reactions_df.merge(posts_user_count_df,
                                                        on="user_id",
                                                        how="left")
user_posts_reactions_df
```

**Now after merging of the posts cound and reactions count we want to make the NaN values to 0 so that we can calculate the sum of the reaction and posts count**

```python
user_posts_reactions_df[['reaction_count', 'posts_count']] = user_posts_reactions_df.loc[:,['reaction_count', 'posts_count']].fillna(value=0)
user_posts_reactions_df
```

**The sum of both the posts count and reaction count for finding their combined effect**

```python
user_posts_reactions_df["added_reaction_post_value"] = user_posts_reactions_df['reaction_count'] + user_posts_reactions_df['posts_count']
user_posts_reactions_df = user_posts_reactions_df.sort_values("added_reaction_post_value")
user_posts_reactions_df
```
